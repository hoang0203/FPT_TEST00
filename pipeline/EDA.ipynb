Ok so If it was a correct email address, It must start without "invalid" word, as well as have a domain. start off '@' letter and then organisation name and '.com', '.net', or '.org' 

```python
def is_valid_email(email):
    if pd.isnull(email):
        return False
    email = str(email)
    if email.lower().startswith('invalid'):
        return False
    if '@' not in email:
        return False
    username, _, domain = email.partition('@')
    if not username.strip():
        return False
    if '.' not in domain:
        return False
    if not domain.lower().endswith(('.com', '.net', '.org')):
        return False
    return True

filtered_customers_df = customers_df[customers_df['email'].apply(is_valid_email)]
```

```python
transactions_df = pd.read_csv(r'D:\Project\Test\FPT_TEST00\datasource\transactions.csv')
```

```python
# remove duplicate entries
transactions_df = transactions_df.drop_duplicates()
# find invalid transaction dates (e.g., 2024-13-45).
def is_valid_date(date_str):
    for fmt in ('%m/%d/%Y', '%Y-%m-%d'):
        try:
            pd.to_datetime(date_str, format=fmt, errors='raise')
            return True
        except Exception:
            continue
    return False

valid_transactions_df = transactions_df[transactions_df['transaction_date'].apply(is_valid_date)]
```

```python
valid_transactions_df.info()
```

```python
valid_transactions_df['transaction_id_LEN'] = valid_transactions_df['transaction_id'].str.len()
valid_transactions_df['customer_id_LEN'] = valid_transactions_df['customer_id'].str.len()
valid_transactions_df['transaction_date_LEN'] = valid_transactions_df['transaction_date'].str.len()

valid_transactions_df.describe(include='all')
```
